```python
# this data had a link that was easily imported in pandas withot importing the csv file
df_power = pd.read_csv('https://query.data.world/s/fow7pwzldz7lqojt6iruxnbjghnnvj')
df_power 
```

```python
# dropped the empty rows
df_power_clean= df_power.dropna()
df_power_clean
```

```python
# rename the colomns with shorter names and lower letters
df_power_rename= df_power_clean.rename(columns= {
    "Country Name":"country",
    "Country Code":"country_code",
    "Year":"year",
    "Proportion of seats held by women in national parliaments (%)":"parliaments_seats",
})
df_power_rename
```

```python
# muliplied by 100 the percentage colomn to look more appeling
df_power_rename["parliaments_seats"]= 100*df_power_rename["parliaments_seats"]
df_power_rename
```

```python
#save to csv in the Data folder
df_power_rename.to_csv("Data/parliaments_seats.csv")
```

```python
# import the csv with companies female managers
df2="Resources/companies_with_female_managers.csv"

#Read files and store in pandas
df_firms = pd.read_csv(df2)
df_firms.head()
```

```python
# fill the NaN with "0" value
df_firms_fill= df_firms.fillna(0)
df_firms_fill.head()
```

```python
#remove additional columns in this case the Indicator Code
df_firms_short = df_firms_fill.drop(["Indicator Code"], axis=1)
df_firms_short
```

```python
# remove colomn years between years 1960 to 2018
df_companies= df_firms_short.drop(df_firms_short.iloc[:,3 :52], axis=1)
df_companies
```

```python
# rename the colomns with shorter names and lower letters
df_companies_rename= df_companies.rename(columns= {
    "Country Name":"country",
    "Country Code":"country_code",
    "Indicator Name":"indicator_name",
})
df_companies_rename
```

```python
#save to csv in the Data folder
df_companies.to_csv("Data/top_manager.csv")
```
